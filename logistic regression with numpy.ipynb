```python
# creating the dataset
import numpy as np
x = np.array([[1,2], [1,3], [2,1], [2,3], [2,4], [3,1], [3,2], [3,4], [4,2], [4,3]])
y = np.array([1, 1, 0, 1, 1, 0, 0, 1, 0, 0])
x1 = x[:, 0].reshape(10, 1)
x2 = x[:, 1].reshape(10, 1)
```

```python
# visualizing the dataset
import matplotlib.pyplot as plt
plt.scatter(x1[y == 1], x2[y == 1], c = 'blue')
plt.scatter(x1[y == 0], x2[y == 0], c = 'red')
plt.title('classification')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python
# implement gradient descent algorithm
X = np.hstack((np.ones((x.shape[0], 1)), x))
y = y.reshape(10, 1)
W = np.zeros((X.shape[1], 1))
alpha = 1
m = len(X)
costs= []
for i in range(200):
    y_pred = sigmoid(np.matmul(X, W))
    # calculate the cost function
    cost = (-1/m)*np.sum(y*np.log(y_pred) + (1 - y)*np.log(1 - y_pred))
    costs.append(cost)
    # calculate the gradient
    grad = (1/m)*np.matmul(X.T, (y_pred - y))
    # updating the parametries or the weight
    W = W - alpha*grad
    if i%10 == 0:
        print('the cost at', i ,'iteration is', cost)
```

```python
# visualizing the cost function
plt.plot(range(200), costs)
plt.title('cost vs iteration')
plt.xlabel('iterations')
plt.ylabel('cost')
plt.show()
```

```python
# creating and visualizing the decision boundery
x_2 = -(W[0] + W[1]*x1)/W[2]
plt.scatter(x1[y == 1], x2[y == 1], c = 'blue')
plt.scatter(x1[y == 0], x2[y == 0], c = 'red')
plt.plot(x1, x_2)
plt.title('classification')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```

```python
# sample prediction
def predict(x1, x2):
    X = [1, x1, x2]
    y_pred = sigmoid(np.matmul(X, W))
    if y_pred >= 0.5:
        return 1
    else:
        return 0
```
