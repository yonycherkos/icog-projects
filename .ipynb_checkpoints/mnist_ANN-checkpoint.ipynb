# classifing mnist hand written digits

## data preprocessing

```python
# importing the libraries
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense
import matplotlib.pyplot as plt
from keras.utils import to_categorical
```

```python
# importing the dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
# visualize sample of the dataset
print(y_train[5000])
plt.imshow(X_train[5000], cmap = 'Greys')
X_train.shape
```

```python
# reshape the image
image_size = 784
num_classes = 10
X_train = X_train.reshape(X_train.shape[0], image_size)
X_test = X_test.reshape(X_test.shape[0], image_size)
X_train.shape
```

```python
# convet the output into binary output
y_train = to_categorical(y_train, num_classes)
y_test = to_categorical(y_test, num_classes)
```

## Building The Neural Network

```python
# creating model and adding layers to it
model = Sequential()
model.add(Dense(units = 32, activation = 'sigmoid', input_shape = (image_size, ))) # the input layer and the hidden layer
model.add(Dense(units = num_classes, activation = 'softmax')) # the output layer
model.summary()
```

## Training and Evaluating The Model

```python
# compling and fitting the train data to the model
model.compile(optimizer = 'sgd', loss = 'categorical_crossentropy', metrics = ['accuracy'])
history = model.fit(X_train, y_train, batch_size=128, epochs=5, validation_split = .1)
loss, accuracy = model.evaluate(X_test, y_test)
```

```python
# visuallizing the train and evaluation accuracy
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training', 'validation'], loc='best')
plt.show()

# display the loss and accuracy for the test set
print(f'test loss: {loss:.3}')
print(f'accuracy: {accuracy:.3}')
```

## Expermenting with different numbers of layers and widths

```python
# Creating the experimenting model
def creat_dense(layer_size):
  
  model = Sequential()
  
  model.add(Dense(units = layer_size[0], activation = 'sigmoid', input_shape = (image_size, ))) # the input and the first hidden layer
  # the rest hidden layers
  for size in layer_size[1:]:
    model.add(Dense(units = size, activation = 'sigmoid', input_shape = (image_size, )))
    
  model.add(Dense(units = number_class, activation = 'softmax')) # outpuy layer
  return model
```

```python
# Evalutating the experimental model
def evaluate(model, batch_size = 128, epochs = 5):
  
  model.compile(optimizer = 'sgd', loss = 'categorical_crossentropy', metrics = ['accuracy'])
  
  history = model.fit(X_train, y_train, batch_size = batch_size, epochs = epochs, validation_split = .1, verbose = False)
  loss, accuracy = model.evaluate(X_test, y_test, verbose = False) 
                                                                                 
  plt.plot(history.history['acc'])
  plt.plot(history.history['val_acc'])
  plt.title('model accuracy')
  plt.xlabel('epochs')
  plt.ylabel('accuracy')
  plt.legend(['training', 'evaluation'], loc = 'best')
  plt.show()
                                                                                 
  print(f'Test Loss: {loss:.3}')
  print(f'accuracy: {accuracy:.3}')                                                                               
```

###  compare longer chains

```python
for layer in range(1, 5):
  model = creat_dense([32] * layer)
  evaluate(model)
  print('#' * 80)
```

### comparing numbers of nodes per layer

```python
for node in [32, 64, 128, 256, 512]:
  model = creat_dense([node])
  evaluate(model)
  print('#' * 80)
```
